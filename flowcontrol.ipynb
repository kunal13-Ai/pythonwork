```python
a = int(input("enter score:")) # Ensure 'a' is an integer
factorial = 1
if a < 0:
  print("Factorial does not exist for negative numbers")
elif a == 0:
  print("The factorial of 0 is 1")
else:
  for i in range(1, a + 1): # Iterate from 1 up to and including 'a'
    factorial = factorial * i # Correct factorial calculation
  print("The factorial of", a, "is", factorial) # Print final result
```
