```python
def factorial(num):
  result=1
  while num>=1:
    result = result*num
    num=num-1
  return result
```

```python
factorial(6)
```

```python
# Factorial
def fact(num):
  result=1
  while num >= 1:
    result=result*num
    num=num-1
  return result
```

```python
fact(5)
```

```python
def sum_sub(x,y):
  sum=x+y
  sub=x-y
  return sum, sub
```

```python
x,y=sum_sub(100,50)
print("The Sum is",x)
print("The sub is",y)
```

```python
#fibonacci = its sum of pervious 2 numbers
def fibonacci(n):
  if n<=1:
   return n
  return fibonacci(n-1)+fibonacci(n-2) # Recursive function
```

```python
print(fibonacci(5))
print(fibonacci(9))
```

```python
num=5
fact=1
for i in range(num):
  i=i+1
  fact=fact*i
print(fact)
```

```python
# using while loop

num=5
i=1
while i>=num:
  i=i+1
  fact=i*fact
print(fact)
```
