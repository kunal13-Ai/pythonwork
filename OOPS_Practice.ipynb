```python
class Atm :

  # constructor
  def __init__(self,owner, balance=0):
    self.pin     = ''
    self.balance = balance
    self.owner = owner
    self.menu()

  def menu(self):
    user_input = input("""
    hi, can I help you?
    1. press 1 to change pin
    2. press 2 to create pin
    3. press 3 to check balance
    4. press 4 to withdraw money
    5. press 5 to deposit money
    6. anything else to exit
    """)

    if user_input == '1':
      self.create_pin()
      pass
    elif user_input == '2':
      self.change_pin()
      pass
    elif user_input == '3':
      self.check_bal()
      pass
    elif user_input =='4' :
      self.withdraw()
      pass
    elif user_input == '5':
      self.deposit()
      pass
    else:
      exit()
      print("Thank you , please visit again")

  def create_pin(self):
   user_pin = input("enter your pin")
   self.pin = user_pin
   print('pin created sucessfully')
   self.menu()

  def change_pin(self):
    old_pin = input('enter old pin')

    if old_pin == self.pin :
     new_pin = input('enter new pin')
     self.pin = new_pin
     print('pin change sucsseful')

    else :
      print('wrong , enter correct pin')
    self.menu()

  def check_bal(self):
    user_pin = input('enter your pin')

    if user_pin == self.pin :
      print('your balance is', self.balance)

    else :
      print('wrong pin, please enter correct pin')
    self.menu()

  def withdraw(self):
    user_pin = input('enter your pin')

    if user_pin == self.pin :
      amount = int(input('enter amount'))

      if amount <= self.balance :
       self.balance = self.balance - amount
       print('withdraw sucsessful')
       print('After withdraw',amount,'your remaining balance is',self.balance)

      else:
        print('Insufficient balance')

    else :
      print('Wrong pin please enter correct pin')
    self.menu()

  def deposit(self):
    user_pin = input('enter your pin ')
    if user_pin == self.pin :
      amount = int(input('enter amount for deposit'))
      self.balance =+ amount
      print('After deposit your balance is',self.balance)
      self.menu()
    else :
      print('sorry')
    self.menu()
```
